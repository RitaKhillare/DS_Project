```python
url = "https://www.opindia.com/latest-news/"
A = requests.get(url)
B = A.content
soup = bs4.BeautifulSoup(B,'html.parser')
```

```python
URLS = []
for i in soup.find_all("h3"):
    URLS.append(i.find("a")["href"])
```

```python
articles = []
HLs = []
dates = []
for i in URLS:
    url = i
    A = requests.get(url)
    B = A.content
    soup = bs4.BeautifulSoup(B,'html.parser')
    headlines = soup.find("h1").text
    Date_of_article = soup.find("time").text
    art = ""
    for j in soup.find_all("p"):
        art = art + j.text
    articles.append(art)
    HLs.append(headlines)
    dates.append(Date_of_article)
```

```python
import pandas as pd
W = pd.DataFrame([dates,HLs,articles]).T
W.columns = ["Date","Headline","Article"]
```
